```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
```

```python
register_matplotlib_converters()
```

```python
fcc_forum_pageview = pd.read_csv("G:\\Otros ordenadores\\Mi Portátil\\Gaspar\\Analisis de datos\\Free code camp\\Page View Time Series Visualizer\\fcc-forum-pageviews.csv")

fcc_forum_pageview
```

```python
fcc_forum_pageview.info()
```

**Cleaning of erroneous data**

```python
# Elimination of elements with values in the 2.5% superior or inferior range

fcc_forum_pageview = fcc_forum_pageview[fcc_forum_pageview["value"] >= fcc_forum_pageview["value"].quantile(0.025)]

fcc_forum_pageview = fcc_forum_pageview[fcc_forum_pageview['value'] <= fcc_forum_pageview['value'].quantile(0.975)]
```

```python
# Transformation of elements in the "date" column to date data type

fcc_forum_pageview['date'] = pd.to_datetime(fcc_forum_pageview['date'])
```

**Daily freeCodeCamp Forum Page Views**

```python
# Creation of the function `draw_line_plot` to draw a bar chart with page views data
def draw_line_plot(data, 
                   x_column,
                   y_column,
                   title,
                   x_label,
                   y_label,
                   figzise = (16, 4)):
    fig, axes = plt.subplots(figsize = (16, 4))
    
    plt.plot(fcc_forum_pageview['date'], fcc_forum_pageview["value"])
    plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
    plt.xlabel("Date")
    plt.ylabel("Page Views")

    
```

```python
draw_line_plot(
    data= fcc_forum_pageview,
    x_column= fcc_forum_pageview['date'],
    y_column= fcc_forum_pageview['value'], 
    title= "Daily freeCodeCamp Forum Page Views 5/2016-12/2019",
    x_label= "Page Views",
    y_label= "Date"
)
```

**Average daily visits per month**

```python
# Creation of two new columns for months and years

fcc_forum_pageview["year"] = fcc_forum_pageview['date'].dt.year

fcc_forum_pageview["month"] = fcc_forum_pageview['date'].dt.month
```

```python
fcc_forum_pageview
```

```python
import calendar

month_name = {i: calendar.month_name[i] for i in range(1, 13)}

fcc_forum_pageview["month"] = fcc_forum_pageview["month"].map(month_name)

fcc_forum_pageview
```

```python
fcc_forum_pageview_mean = fcc_forum_pageview.groupby(["year","month"]).mean("value").round()
```

```python
fcc_forum_pageview_mean
```

```python
Average_Page_Views = sns.catplot(x = "year",
                                 y = "value",
                                 hue = "month",
                                 data = fcc_forum_pageview_mean,
                                 kind = "bar",
)

Average_Page_Views.set(xlabel = "Year", ylabel = "Average Page Views")

```

```python
def draw_bar_plot(data,
                  x_column,
                  y_column,
                  hue,
                  kind):
    Average_Page_Views = sns.catplot(x = "year",
                                     y = "value",
                                     hue = "month",
                                     data = fcc_forum_pageview,
                                     estimator = np.mean,
                                     ci = None
)

Average_Page_Views.set(xlabel = "Year", ylabel = "Average Page Views")


```
