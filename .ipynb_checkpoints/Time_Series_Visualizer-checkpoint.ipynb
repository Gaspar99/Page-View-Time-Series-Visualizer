```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
```

```python
register_matplotlib_converters()
```

```python
fcc_forum_pageview = pd.read_csv("G:\\Otros ordenadores\\Mi Portátil\\Gaspar\\Analisis de datos\\Free code camp\\Page View Time Series Visualizer\\fcc-forum-pageviews.csv")

fcc_forum_pageview
```

```python
fcc_forum_pageview.info()
```

**Cleaning Of Erroneous Data**

```python
# Elimination of elements with values in the 2.5% superior or inferior range

fcc_forum_pageview = fcc_forum_pageview[fcc_forum_pageview["value"] >= fcc_forum_pageview["value"].quantile(0.025)]

fcc_forum_pageview = fcc_forum_pageview[fcc_forum_pageview['value'] <= fcc_forum_pageview['value'].quantile(0.975)]
```

```python
# Transformation of elements in the "date" column to date data type

fcc_forum_pageview['date'] = pd.to_datetime(fcc_forum_pageview['date'])
```

**Daily freeCodeCamp Forum Page Views**

```python
# Creation of the function `draw_line_plot` to draw a bar chart with page views data
def draw_line_plot(data, 
                   x_column,
                   y_column,
                   title,
                   x_label,
                   y_label,
                   figzise = (16, 4)):
    fig, axes = plt.subplots(figsize = (16, 4))
    
    plt.plot(fcc_forum_pageview['date'], fcc_forum_pageview["value"])
    plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
    plt.xlabel("Date")
    plt.ylabel("Page Views")

    
```

```python
draw_line_plot(
    data= fcc_forum_pageview,
    x_column= fcc_forum_pageview['date'],
    y_column= fcc_forum_pageview['value'], 
    title= "Daily freeCodeCamp Forum Page Views 5/2016-12/2019",
    x_label= "Page Views",
    y_label= "Date"
)
```

**Average Daily Visits Per Month**

```python
# Creation of two new columns for months and years

fcc_forum_pageview["year"] = fcc_forum_pageview['date'].dt.year

fcc_forum_pageview["month"] = fcc_forum_pageview['date'].dt.month
```

```python
fcc_forum_pageview
```

```python
# Transformation of the months from numbers to month names

import calendar

month_name = {i: calendar.month_name[i] for i in range(1, 13)}

fcc_forum_pageview["month"] = fcc_forum_pageview["month"].map(month_name)

fcc_forum_pageview
```

```python
# Grouping of the data by months and years
# Calculation of the average daily visits


fcc_forum_pageview_mean = fcc_forum_pageview.groupby(["year","month"]).mean("value").round()
```

```python
fcc_forum_pageview_mean
```

```python
# Creation of the function `draw_bar_plot` to generate the chart of daily average visits per month

def draw_bar_plot(data,
                  x_column,
                  y_column,
                  hue,
                  kind):
    plt.figure(figsize=(8, 6))
    
    Average_Page_Views = sns.barplot(x = "year",
                                     y = "value",
                                     hue = "month",
                                     data = fcc_forum_pageview,
                                     estimator = np.mean,
                                     errorbar = None
)

    Average_Page_Views.set(xlabel = "Year", ylabel = "Average Page Views")


```

```python
draw_bar_plot(data = fcc_forum_pageview,
              x_column = "year",
              y_column = "value",
              hue = "month",
              kind = "bar",
              )
```

**Box Plots**

```python
fcc_forum_pageview['cyclic_month'] = (fcc_forum_pageview['date'].dt.month + (fcc_forum_pageview['year'] - fcc_forum_pageview['year'].min()) * 12) % 12 + 1

fcc_forum_pageview.sort_values(by = "cyclic_month")
```

```python
fcc_forum_pageview_cyclic_month = fcc_forum_pageview.sort_values(by = "cyclic_month") 
```

```python
box_plot = plt.subplots(1, 2, figsize = (26,8))

fig, (ax1, ax2) = box_plot
box_plot

sns.boxplot(y = fcc_forum_pageview_cyclic_month["value"],
            x = fcc_forum_pageview_cyclic_month['year'],
            ax = ax1,
            hue = fcc_forum_pageview_cyclic_month["year"]
)

ax1.legend().set_visible(False)

ax1.set_title("Year-wise Box Plot (Trend)")
ax1.set_ylabel("Page View")
ax1.set_xlabel("Year")


sns.boxplot(y = fcc_forum_pageview['cyclic_month'] = (fcc_forum_pageview['date'].dt.month + (fcc_forum_pageview['year'] - fcc_forum_pageview['year'].min()) * 12) % 12 + 1,
            x = fcc_forum_pageview_cyclic_month['cyclic_month'],
            ax = ax2,
            hue = fcc_forum_pageview_cyclic_month['cyclic_month']
) 

plt.xticks(ticks=range(0, 13), labels=["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic", ""])

ax2.set_title("Month-wise Box Plot (Seasonality)")
ax2.set_ylabel("Page View")
ax2.set_xlabel("Month")
```
